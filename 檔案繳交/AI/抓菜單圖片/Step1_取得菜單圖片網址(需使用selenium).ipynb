```python
from selenium import webdriver
import time
```

```python
# 菜單網址 https://www.musthavemenus.com/category/restaurant-menu.html
driver = webdriver.Edge()
driver.get("https://www.musthavemenus.com/category/restaurant-menu.html")
```

```python
for x in range(20):
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(1)
```

```python
menu_name = driver.find_elements_by_class_name("item-content")
print(len(menu_name))
```

```python
menu_link = []
for index, name in enumerate(menu_name):
    temp = name.text.lower()
    temp = temp.replace(" ","-")
    menu_link.append("https://www.musthavemenus.com/menu-template/" + temp + ".html")
    print(menu_link[index])
```

```python
import numpy as np 
np.savetxt("menu_link.csv", menu_link, delimiter =",",fmt ='%s')
```

```python
driver.close()
```
