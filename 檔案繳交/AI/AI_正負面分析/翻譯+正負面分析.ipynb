```python
# 載入相關套件
from transformers import pipeline
```

```python
# 載入模型
classifier = pipeline('sentiment-analysis')
```

```python
from googletrans import Translator

def trans(text):
    translator = Translator()
    result = translator.translate(text = text, dest='en', src='zh-TW').text
    return result
```

```python
text = "防疫外帶-在家也能用麻辣鍋與白鍋，一次兩種湯底的三人火鍋套餐"
print(trans(text))
print(classifier(trans(text)))
```

```python
text = "240太貴，不推薦，普通"
print(trans(text))
print(classifier(trans(text)))
```

```python
text = "飯比雞優，粒粒明，濃厚雞汁香味"
print(trans(text))
print(classifier(trans(text)))
```

```python
text = "比起一些費心思裝修的韓式餐廳，這家把心思用在料理上，真的好吃"
print(trans(text))
print(classifier(trans(text)))
```
